### Task 3: Several stations show a prominent annual periodicity. Identify the 4 **stations** (identify them by the index of their location in the data cube) that show the most prominent periodic trend on an annual period (52 weeks). Can you figure out what the periodic peak in rides is due to?


```python
N = len(sum_station[0])
x_labels = [1.0/194, 1.0/52, 1.0/26, 1.0/8, 1.0/6, 1.0/4, 1.0/2]
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(111)
for i in range(0,len(sum_station)):
    f = np.abs(np.fft.rfft(sum_station[i]))
    ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (f[1:]), '-')

ax.set_xticks(x_labels)
ax.set_xticklabels([ "%s"%(1.0/f) for f in x_labels], fontsize=15, rotation = 70)

plt.xlabel("Period (weeks)", fontsize=12)
plt.ylabel("Power", fontsize=12)
plt.title("Fig 4: Subway Rides Power Spectrum", fontsize=14)
```

#### Figure 4: Shows the power spectrum of each MTA station's time series. There are a few obvious peaks in the graph. We can calculate the power spectrum for each station to find the stations with the strongest annual periodicity.

```python
freq_cum = np.abs(np.fft.rfft(sum_station))
```

```python
annual_power = []
for i in range(0,len(freq_cum)):
    tmp = freq_cum[i][4]
    annual_power.append(tmp)
```

```python
df = pd.DataFrame(annual_power, columns=['Annual_Periodicity'])
df.sort_values(by=['Annual_Periodicity'],inplace = True)
df.tail()
```

#### The above table illustrates that the stations 0, 194, 150, and 327 shows the highest annual periodicity. 

```python
plt.figure(figsize=(20,20)) #plot to understand why periodicity occurs
for i,stat in enumerate([0, 194, 150, 327]):
    plt.subplot(2,2,i+1)
    plt.plot(date, sum_station[stat], 'navy')
    plt.ylabel('Weekly Subway Rides', size=11)
    plt.xlabel('Date', size=11)
    plt.xticks(rotation=75)
    plt.title(('Fig 5: Weekly MTA rides for %sth station from 05/21/2010 - 02/07/2014'%stat), size = 14)
```

#### Figure 5 shows that weekly rides increases in Sept every year and reaches lowest peak in Jan. This might be due to Sept being the start of the new school year when there is increased mobility by students and Jan being a colder month that sees more people using alternative transportation methods.
